```python
import numpy as np 
import matplotlib.pyplot as plt
```

Task 1: Line plot

```python
x = np.linspace(0,10,100) # 100 points from 0 to 10
y = np.sin(x) #sine function 

plt.figure(figsize=(8,4))
plt.plot(x, y, label = 'sin(x)', color= 'red', linestyle="--", linewidth = 2)
plt.title("Line plot of sin(x)")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.linspace(0,10,100) # 100 points from 0 to 10
y = np.cos(x) #sine function 

plt.figure(figsize=(8,4))
plt.plot(x, y, label = 'cos(x)', color= 'red', linestyle="--", linewidth = 2)
plt.title("Line plot of cos(x)")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.legend()
plt.grid()
plt.show()
```

Task 2: Plot a scatter chart

```python
x = np.random.rand(50) # 50 random generated points
y = np.random.rand(50)

colors = np.random.rand(50)
sizes = np.random.rand(50) * 1000 #scaling up the size

plt.figure(figsize=(8,4))
plt.scatter(x, y, c = colors, s = sizes, alpha = 0.5, cmap = 'viridis')
plt.title("Scatter plot")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.legend()
plt.grid()
plt.show()
```

Task 3: Bar chart

```python
categories = ['A','B', 'C', 'D','E','F','G','H']
values= np.random.randint(10, 100, size = 8)
colors = np.random.rand(9)

plt.figure(figsize=(8,4))
plt.bar(categories, values, color = ['red','green','blue','yellow', 'orange','black','pink', 'purple'])
plt.title("Bar chart on random values")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.legend()
plt.grid()
plt.show()
```

Task 4: Histogram

```python
data = np.random.normal(0,1,1000) # normal distribution
plt.figure(figsize=(8,4))
plt.hist(data, bins= 20, color= 'teal', edgecolor= 'black', alpha = 0.8)
plt.title("Histogram of normallt distributed data")
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

Task 5: Pie chart

```python
segments = ['Seg 1','Seg 2','Seg 3','Seg 4','Seg 5' ]
sizes = np.random.randint(10,50, size = 5)
plt.figure(figsize= (12,12))
plt.pie(sizes, labels=segments, autopct="%1.1f%%", startangle= 90, colors=['blue','yellow', 'orange','pink', 'purple'] )
plt.title("Pie chart")
plt.legend(title='Segments')
plt.show()
```
