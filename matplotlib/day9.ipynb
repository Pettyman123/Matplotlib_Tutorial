Task: 3D Surface Plot with Customizations
Visualize a mathematical function using a 3D surface plot. Add the following customizations:
A color map representing the height of the surface.
Labels for the X, Y, and Z axes.
A color bar for reference.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create the figure and 3D axes
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Add labels
ax.set_title("3D Surface Plot of z = sin(sqrt(x^2 + y^2))", fontsize=16)
ax.set_xlabel("X-axis", fontsize=12)
ax.set_ylabel("Y-axis", fontsize=12)
ax.set_zlabel("Z-axis", fontsize=12)

# Add a color bar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label="Height (z)")

# Show plot
plt.show()
```
