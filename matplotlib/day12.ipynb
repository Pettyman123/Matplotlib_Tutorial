Task: Stock Market Analysis Dashboard
Create a stock market visualization to analyze the weekly performance of a fictional stock. Your task will include:

A candlestick-like line plot showing opening, closing, high, and low prices.
A bar chart showing daily trading volume.
A scatter plot comparing daily returns vs. volume.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random stock data
np.random.seed(42)
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
open_prices = np.random.randint(100, 150, size=5)
close_prices = open_prices + np.random.randint(-10, 10, size=5)
high_prices = np.maximum(open_prices, close_prices) + np.random.randint(1, 5, size=5)
low_prices = np.minimum(open_prices, close_prices) - np.random.randint(1, 5, size=5)
volume = np.random.randint(1000, 5000, size=5)
daily_returns = ((close_prices - open_prices) / open_prices) * 100  # Percentage change

# Create figure
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Stock Market Analysis Dashboard", fontsize=16)

# Line Plot: Stock price movements (Candlestick-style)
ax1 = axs[0, 0]
ax1.plot(days, open_prices, label="Open", marker="o", linestyle="--", color="blue")
ax1.plot(days, close_prices, label="Close", marker="s", linestyle="-", color="red")
ax1.fill_between(days, low_prices, high_prices, color="gray", alpha=0.3, label="High-Low Range")
ax1.set_title("Stock Price Movements")
ax1.set_xlabel("Day")
ax1.set_ylabel("Stock Price ($)")
ax1.legend()
ax1.grid()

# Bar Chart: Daily Trading Volume
ax2 = axs[0, 1]
ax2.bar(days, volume, color="purple", alpha=0.7, label="Volume")
ax2.set_title("Daily Trading Volume")
ax2.set_xlabel("Day")
ax2.set_ylabel("Volume (Shares)")
ax2.legend()

# Scatter Plot: Daily Returns vs. Volume
ax3 = axs[1, 0]
scatter = ax3.scatter(daily_returns, volume, c=daily_returns, cmap="coolwarm", edgecolor="k", s=100)
ax3.set_title("Daily Returns vs. Trading Volume")
ax3.set_xlabel("Daily Returns (%)")
ax3.set_ylabel("Volume (Shares)")
fig.colorbar(scatter, ax=ax3, label="Daily Return (%)")

# Empty subplot for alignment
axs[1, 1].axis("off")

# Adjust layout and display
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
