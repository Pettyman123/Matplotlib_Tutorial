Task: Climate Change Analysis (Temperature & CO₂ Levels Over Time)
Create a climate change dashboard that visualizes global temperature rise and CO₂ emissions over time. Your task will include:

A dual-axis line plot showing global average temperature increase and CO₂ levels.
A scatter plot comparing temperature anomalies with CO₂ emissions.
A heatmap-style bar chart for seasonal temperature variations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random climate data
np.random.seed(42)
years = np.arange(2000, 2021)
global_temp = np.linspace(0.2, 1.5, len(years)) + np.random.uniform(-0.1, 0.1, len(years))  # Temperature anomaly (°C)
co2_levels = np.linspace(370, 420, len(years)) + np.random.uniform(-2, 2, len(years))  # CO2 ppm
season_labels = ["Winter", "Spring", "Summer", "Fall"]
season_temp = np.random.uniform(-5, 10, size=(len(years), 4))

# Create figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Climate Change Analysis Dashboard", fontsize=16)

# Dual-axis Line Plot: Global Temperature & CO₂ Levels
ax1 = axs[0, 0]
ax1.plot(years, global_temp, label="Temperature Anomaly (°C)", color="red", marker="o")
ax2 = ax1.twinx()
ax2.plot(years, co2_levels, label="CO₂ Levels (ppm)", color="blue", linestyle="--", marker="s")
ax1.set_xlabel("Year")
ax1.set_ylabel("Temperature Anomaly (°C)", color="red")
ax2.set_ylabel("CO₂ Levels (ppm)", color="blue")
ax1.set_title("Global Temperature vs. CO₂ Levels")
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

# Scatter Plot: Temperature Anomalies vs. CO₂ Emissions
ax3 = axs[0, 1]
scatter = ax3.scatter(co2_levels, global_temp, c=global_temp, cmap="coolwarm", edgecolor="black")
ax3.set_xlabel("CO₂ Levels (ppm)")
ax3.set_ylabel("Temperature Anomaly (°C)")
ax3.set_title("Temperature Anomalies vs. CO₂ Levels")
fig.colorbar(scatter, ax=ax3, label="Temperature Anomaly (°C)")

# Heatmap-style Bar Chart: Seasonal Temperature Variations
ax4 = axs[1, 0]
for i, season in enumerate(season_labels):
    ax4.bar(years, season_temp[:, i], label=season, alpha=0.7)
ax4.set_title("Seasonal Temperature Variations")
ax4.set_xlabel("Year")
ax4.set_ylabel("Temperature Change (°C)")
ax4.legend()

# Empty subplot for alignment
axs[1, 1].axis("off")

# Adjust layout and display
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
