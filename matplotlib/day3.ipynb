```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

Task 1: Advanced Subplots with Shared Axes

Create a 2x2 grid of subplots:
Line plot (y=sin(x)).
Scatter plot (y=cos(x)).
Bar plot with random values.
Histogram of normally distributed random data.
Share the x-axis across all subplots.

```python
x = np.linspace(0,10,100)
fig, axs = plt.subplots(2,2, figsize=(8,12), sharex = True)

#line plot code
axs[0,0].plot(x, np.sin(x), color= 'blue', label='y = sinx')
axs[0,0].set_title('Line plot')
axs[0,0].legend()

#scatter plot code
axs[0,1].scatter(x, np.cos(x), color='green', alpha =0.7, label= 'y = cosx')
axs[0,1].set_title('Scatter plot')
axs[0,1].legend()

#bar plot code
random_values = np.random.randint(1, 100, size=5)
axs[1,0].bar(range(len(random_values)), random_values, color= 'orange')
axs[1, 0].set_title('Bar Plot')

#histogram plot code
random_data = np.random.randn(1000)
axs[1,1].hist(random_data, bins=20, color='purple', alpha=0.5)
axs[1,1].set_title('Histogram')

plt.tight_layout()
plt.show()
```

Task 2: Time-Series Data Visualization

Generate random time-series data for the last 30 days.
Plot the data as:
A line plot for values.
A bar chart showing daily differences.
Add grid, title, and proper formatting for dates.

```python
dates = pd.date_range(end=pd.Timestamp.today(), periods=30)
values = np.random.randint(50, 150, size=30)
differences = np.diff(values, prepend=values[0])

fig, ax = plt.subplots(2, 1, figsize=(10, 8))

# Line plot
ax[0].plot(dates, values, marker='o', linestyle='-', label='Values')
ax[0].set_title('Time-Series Data')
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Value')
ax[0].grid(True)
ax[0].legend()

# Bar chart for daily differences
ax[1].bar(dates, differences, color='teal', alpha=0.7, label='Daily Difference')
ax[1].set_title('Daily Differences')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Difference')
ax[1].grid(True)
ax[1].legend()

plt.tight_layout()
plt.show()
```

Task 3: Real-World Dataset (Iris Dataset)

Load the Iris dataset using seaborn or sklearn.
Plot the following:
A scatter plot showing the relationship between petal length and petal width for different species.
A box plot comparing sepal lengths across species.

```python
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Scatter plot
plt.figure(figsize=(8, 6))
species_colors = ['red', 'blue', 'green']
for species, color in enumerate(species_colors):
    subset = df[df['species'] == species]
    plt.scatter(subset['petal length (cm)'], subset['petal width (cm)'],
                label=iris.target_names[species], color=color)

plt.title('Petal Length vs. Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(8, 6))
df.boxplot(column='sepal length (cm)', by='species', grid=False, patch_artist=True,
           boxprops=dict(facecolor='lightblue', color='blue'))
plt.title('Sepal Length by Species')
plt.suptitle('')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.xticks(ticks=[1, 2, 3], labels=iris.target_names)
plt.show()
```

Task 4: Custom Colormap with Contour Plot

Generate a mesh grid with x,y∈[−5,5].
Plot the function z=sin( root(x^2 +y^2 ))
using a contour plot with a custom colormap.

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, Z, cmap='viridis', levels=20)
plt.colorbar(contour, label='Value')
plt.title('Contour Plot with Custom Colormap')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Task 5: Pie Chart with Annotations

Generate random data for 5 categories.
Plot a pie chart with:
Percentage annotations on each slice.
A legend outside the plot.

```python
categories = ['A', 'B', 'C', 'D', 'E']
data = np.random.randint(10, 100, size=5)
explode = [0.1 if max(data) == value else 0 for value in data]

plt.figure(figsize=(8, 6))
plt.pie(data, labels=categories, explode=explode, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart with Annotations')
plt.legend(categories, loc='best')
plt.show()
```
