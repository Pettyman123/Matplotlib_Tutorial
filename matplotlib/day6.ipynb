```python
import numpy as np 
import matplotlib.pyplot as plt
```

Task 1: Overlay multiple line plots on the same axes

```python
x = np.linspace(0,10,100)
y1 = np.tan(x)
y2 = np.sin(x)
y3 = 0.25*y1 + 0.15*y2

plt.plot(x, y1, label='tanx', color='#deb3ad', linestyle='-')
plt.plot(x, y2, label='sinx', color='red', linestyle='--')
plt.plot(x, y3, label='tanx corelate sinx', color='green', linestyle=':')
plt.title("Multiple wave plot")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

Task 2: Stacked Bar chart

```python
quarters=["Q1", "Q2", "Q3", "Q4", "Q5"]
product_a = [1,2,3,4,5]
product_b = [6,7,8,9,10]
product_c = [10,20,30,40,50]

product_b_stack = np.add(product_a, product_b)
product_c_stack = np.add(product_b_stack, product_c)

plt.bar(quarters, product_a, label="Product A", color='#b95c50')
plt.bar(quarters, product_b,bottom=product_a, label="Product B", color='#4297a0')
plt.bar(quarters, product_c,bottom=product_b_stack, label="Product C", color='#6aabd2')

plt.title("Quaterly sales")
plt.xlabel("quarter")
plt.ylabel("sales")
plt.legend()
plt.show()
```

Task 3: Annotating points on scatter plot

```python
np.random.seed(42)
# np.random.seed(42)
x = np.random.rand(50) * 100
y = np.random.rand(50) * 100

indices = np.argsort(y)[-5:] #top five head

plt.scatter(x,y, color='#6aabd2',alpha=0.9, label='Points')
for i in indices:
    plt.annotate(f"({x[i]: .1f}, {y[i]: .1f})", (x[i], y[i]), textcoords='offset points', xytext=(5,5), color='red')

plt.title("Quaterly sales")
plt.xlabel("quarter")
plt.ylabel("sales")
plt.legend()
plt.show()   
```
