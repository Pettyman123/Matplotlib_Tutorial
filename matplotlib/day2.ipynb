```python
import numpy as np 
import matplotlib.pyplot as plt
```

Task 1: Subplots with line and scatter plot
1. A line plot showing the function y = x^2
2. A scatter plot with random x and y values

```python
x = np.linspace(0,20,100)
y1 = x ** 2
x2 = np.random.rand(50)
y2 = np.random.rand(50)

plt.figure(figsize=(8,4))
plt.plot(x, y1, color='blue', label='y = x^2')
plt.title("line plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid()
#plt.show()

plt.subplot(1,2,1)
plt.scatter(x2, y2, color='red', alpha=0.7, label= 'Random points')
plt.title('scatter plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid()
plt.show()
```

Task 2: Customized bar chart
1. generate 5 categories and values
2. plot a horizontal bar with + Customized colors for each bar + Annotations displaying the values at the end of each bar.

```python
categories = ['A','B', 'C', 'D','E']
values= np.random.randint(10, 100, size = 5)
colors = ['red', 'blue', 'green', 'orange', 'pink']

plt.figure(figsize=(8,4))
plt.barh(categories, values, color = colors)
for i, v in enumerate(values):
    plt.text(v+2, i, str(v), color= 'black', fontsize=10, va='center')

plt.title('Horizontal bar chart with annotations')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

Task 3: Multiple Lines on the Same Plot
1. y = x
2. y = x^2
3. y = x^3

```python
x = np.linspace(0, 10, 100)
y1 = x
y2 = x**2
y3 = x**3

plt.figure(figsize=(8,4))
plt.plot(x, y1, label='y1=x', linestyle = '-', color='red')
plt.plot(x, y2, label='y2=x**2', linestyle = '--', color='green')
plt.plot(x, y3, label='y3=x**3', linestyle = ':', color='blue')
plt.title("Multiple functions on the same plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid()
plt.show()
```

Task 4: Advanced Histogram
1. Generate two sets of random data:
Normally distributed (=0,=1).
Uniformly distributed.
2. Plot a stacked histogram comparing the two datasets

```python
data1 = np.random.normal(0,1,1000)
data2 = np.random.uniform(-3,3,1000)

plt.figure(figsize=(8,4))
plt.hist([data1, data2], bins=20, stacked= True, label= ['Normal','Uniform'])
plt.title('Stacked histogram')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

Task 5: Heatmap

1. Generate a 6x6 random matrix.
2. Plot it as a heatmap with:
A color bar.
Custom tick labels for rows and columns.

```python
matrix = np.random.rand(6,6)
row_labels = ['Row1', 'Row2', 'Row3', 'Row4', 'Row5', 'Row6']
col_labels = ['Col1', 'Col2', 'Col3', 'Col4', 'Col5', 'Col6']

plt.figure(figsize=(8,4))
plt.imshow(matrix, cmap='cool', interpolation='nearest')
plt.colorbar(label='value scale')
plt.xticks(ticks= np.arange(6), labels = col_labels)
plt.yticks(ticks=np.arange(6), labels=row_labels)
plt.title('Heatmap')
plt.show()
```
