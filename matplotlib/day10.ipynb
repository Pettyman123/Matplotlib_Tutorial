Task: Sales Dashboard
Create a sales dashboard for a fictional company

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
months = np.arange(1, 13)
regions = ["North", "South", "West"]
categories = ["Electronics", "Furniture", "Clothing"]

# Monthly revenue by region
np.random.seed(42)
north_revenue = np.random.randint(200, 500, size=12)
south_revenue = np.random.randint(150, 400, size=12)
west_revenue = np.random.randint(100, 300, size=12)
```

```python
# Sales contributions by category
category_sales = np.random.randint(500, 1000, size=(12, len(categories)))

# Total yearly revenue by region
region_totals = [north_revenue.sum(), south_revenue.sum(), west_revenue.sum()]
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Sales Dashboard", fontsize=16)
# Line Plot (Monthly revenue by region)
ax1 = axs[0, 0]
ax1.plot(months, north_revenue, label="North", marker="o", linestyle="-")
ax1.plot(months, south_revenue, label="South", marker="s", linestyle="--")
ax1.plot(months, west_revenue, label="West", marker="^", linestyle=":")
ax1.set_title("Monthly Revenue by Region")
ax1.set_xlabel("Month")
ax1.set_ylabel("Revenue ($)")
ax1.legend()
ax1.grid()

# Stacked Bar Chart (Sales contributions by category)
ax2 = axs[0, 1]
width = 0.6
x = np.arange(12)
for i, category in enumerate(categories):
    bottom = category_sales[:, :i].sum(axis=1) if i > 0 else 0
    ax2.bar(x, category_sales[:, i], width, label=category, bottom=bottom)
ax2.set_title("Monthly Sales Contributions by Category")
ax2.set_xlabel("Month")
ax2.set_ylabel("Sales ($)")
ax2.legend()

# Pie Chart (Yearly revenue distribution by region)
ax3 = axs[1, 0]
ax3.pie(region_totals, labels=regions, autopct="%1.1f%%", startangle=140, colors=["blue", "orange", "green"])
ax3.set_title("Yearly Revenue Distribution by Region")

# Empty subplot for alignment
axs[1, 1].axis("off")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
