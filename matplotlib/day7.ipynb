```python
import numpy as np 
import matplotlib.pyplot as plt
```

Task 1: Custom subplot Layout for visualizing trends
==============================================
Top left- line plot showing sine and cosine waves on the same axes
Top right- horizontal bar chart showing a comparison of five categories
Bottom - Scatter plot with a color map representing data intensity

```python
x = np.linspace(0,2* np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

categories = ["Q1", "Q2", "Q3", "Q4", "Q5"]
values = [50,40,60,80,10]

scatter_x = np.random.rand(100)
scatter_y = np.random.rand(100)
scatter_intensity = np.random.rand(100)

fig = plt.figure(figsize=(10,12))
ax1 = plt.subplot2grid((3,3), (0,0), colspan=2)#top left
ax2 = plt.subplot2grid((3,3), (0,2), colspan=2)#top right
ax3 = plt.subplot2grid((3,3), (1,0), colspan=2)#bottom

ax1.plot(x, y_sin, color='blue', linestyle='-')
ax1.plot(x, y_cos, color='red', linestyle='--')
ax1.set_title('Sine and Cosine')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()
ax1.grid()

ax2.barh(categories, values, color='green')
ax2.set_title("category comparison")
ax2.set_xlabel("value")
ax2.set_ylabel("category")
ax2.grid()

sc = ax3.scatter(scatter_x, scatter_y,c = scatter_intensity, cmap='viridis', s =100, alpha=0.7)
ax3.set_title("Scatter plot with intensity")
ax3.set_xlabel("x")
ax3.set_ylabel("y")
fig.colorbar(sc, ax=ax3, orientation="vertical", label='intensity')

plt.tight_layout()
plt.show()
```

Task 2: VIsualize the cluster with centroids

```python
from sklearn.datasets import make_blobs

n_clusters= 4
X, y = make_blobs(n_samples=200, centers=n_clusters, cluster_std=1.0, random_state=42)

centroids= np.array([X[y==i].mean(axis=0) for i in range(n_clusters)])

plt.figure(figsize=(10,10))
for i in range(n_clusters):
    plt.scatter(X[y==i][:,0], X[y==i][:,1], label=f"Cluster {i+1}", alpha=0.6)

plt.scatter(centroids[:,0], centroids[:,1], s= 200, color='black', label = 'Centroids', marker="X")

plt.title("cluster visuals")
plt.legend()
plt.grid()
plt.show()
```
