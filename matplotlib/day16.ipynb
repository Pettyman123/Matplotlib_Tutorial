Task: Earthquake Data Visualization
Create a global earthquake analysis dashboard using real-world-style data. Your task includes:

A scatter plot on a world map showing earthquake locations and magnitudes.
A histogram of earthquake magnitudes to observe frequency distribution.
A time series line plot of earthquake occurrences over a year.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Generate random earthquake data
np.random.seed(42)
num_earthquakes = 100
latitudes = np.random.uniform(-90, 90, num_earthquakes)  # Random latitudes
longitudes = np.random.uniform(-180, 180, num_earthquakes)  # Random longitudes
magnitudes = np.random.uniform(3.0, 9.0, num_earthquakes)  # Magnitudes (Richter scale)
dates = pd.date_range(start="2023-01-01", periods=num_earthquakes, freq="3D")  # Random occurrences over the year

# Create figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Global Earthquake Analysis", fontsize=16)

# Scatter Plot: Earthquake Locations (Magnitude as Size)
ax1 = axs[0, 0]
sc = ax1.scatter(longitudes, latitudes, c=magnitudes, cmap="hot", s=magnitudes**2, alpha=0.7, edgecolors="black")
ax1.set_xlabel("Longitude")
ax1.set_ylabel("Latitude")
ax1.set_title("Global Earthquake Locations")
fig.colorbar(sc, ax=ax1, label="Magnitude (Richter Scale)")

# Histogram: Earthquake Magnitude Distribution
ax2 = axs[0, 1]
ax2.hist(magnitudes, bins=10, color="purple", edgecolor="black", alpha=0.7)
ax2.set_xlabel("Magnitude")
ax2.set_ylabel("Frequency")
ax2.set_title("Distribution of Earthquake Magnitudes")

# Time Series Plot: Earthquake Occurrences Over Time
ax3 = axs[1, 0]
dates_sorted = np.sort(dates)
cumulative_earthquakes = np.arange(1, num_earthquakes + 1)
ax3.plot(dates_sorted, cumulative_earthquakes, marker="o", linestyle="-", color="blue")
ax3.set_xlabel("Date")
ax3.set_ylabel("Cumulative Earthquakes")
ax3.set_title("Earthquake Occurrences Over Time")
ax3.grid()

# Empty subplot for alignment
axs[1, 1].axis("off")

# Adjust layout and display
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
