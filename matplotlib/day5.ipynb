```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error
```

Task 1: Compare Performance of Regression Models
Use synthetic data to compare the performance of Linear Regression and Polynomial Regression models. Visualize the data points and model predictions on the same plot.

```python
np.random.seed(42)
X = np.random.rand( 50,1) * 10
y = 3*X **2 + X * 2 + np.random.randn(50,1)*10

#linear model
model = LinearRegression()
model.fit(X,y)
y_pred = model.predict(X)

#polynomial mode
p_model = PolynomialFeatures(degree=2)
X_poly = p_model.fit_transform(X)
p_model = LinearRegression()
p_model.fit(X_poly,y)
y_pred_poly = p_model.predict(X_poly)

#plot
plt.scatter(X,y, color='blue', label='data points')
plt.plot(X,y_pred, color='red', label='linear model')
plt.plot(X, y_pred_poly, color='green', label='polynomial model')
plt.xlabel('X')
plt.ylabel('y_pred')
plt.legend()
plt.show()
```

Task 2: Plot Clustering Results
Use KMeans clustering on a 2D dataset and visualize the clusters with distinct colors and cluster centroids.

```python
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans

A, b = make_blobs(n_samples=500, centers=5, cluster_std=1.3, random_state=42)
kmeans =KMeans(n_clusters=5, random_state=42)
kmeans.fit(A)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.scatter(A[:,0], A[:, 1], c=labels, cmap='viridis', marker='o', label='Data points')
plt.scatter(centroids[:,0], centroids[:,1], color='red', marker='x',s=100, label='centroid' )
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KMeans Clustering')
plt.legend()
plt.show()
```

Task 3: Feature Importance in Decision Trees
Train a Decision Tree on the Iris dataset and visualize the importance of each feature as a bar chart.

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
T,v = iris.data, iris.target
feature_names = iris.feature_names

tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(T,v)

importance = tree_model.feature_importances_

plt.barh(feature_names, importance, color='teal')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Decision tree')
plt.show()
```
