**Task: weather data visualization dashboard 
Use line plot, bar chart, scatter plot for temperature variations, precipitation levels, humidity vs wind speed respectively

```python
import numpy as np
import matplotlib.pyplot as plt 

np.random.seed(42)

days = ["Monday", "Tuesday", "Wednesday", "Thrusday", "Friday", "Saturday", "Sunday"]
high_temps = np.random.randint(25,35, size=7)
low_temps = np.random.randint(5,15, size=7)
precipitation = np.random.uniform(5,100, size=7)
humidity = np.random.uniform(5,100, size=7)
wind_speed = np.random.uniform(5,100, size=7)
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,10))
fig.suptitle("Dashboard", fontsize = 18)

ax1 = axs[0,0]
ax1.plot(days, high_temps, label="H temp", color='red')
ax1.plot(days, low_temps, label = 'L temp', color='blue')
ax1.legend()
ax1.grid()

ax2 = axs[0,1]
ax2.bar(days, precipitation, color='cyan', alpha=0.65, label ='precipitation')
ax2.legend()

ax3 = axs[1,0]
scatter = ax3.scatter(humidity, wind_speed, c=humidity, cmap="coolwarm", edgecolor='k',s =100)
fig.colorbar(scatter, ax=ax3, label ='humidity')

ax4 = axs[1,1]
ax4.axis("off")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
