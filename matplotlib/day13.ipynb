Task: COVID-19 Cases & Vaccination Analysis
Create a COVID-19 dashboard that visualizes the trend of cases and vaccinations over a period. Your task will include:

A line plot showing daily new cases and cumulative cases.
A stacked area chart representing the percentage of vaccinated vs. unvaccinated population.
A bar chart comparing vaccination rates among different age groups.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random COVID-19 data
np.random.seed(42)
days = np.arange(1, 11)  # 10 days of data
new_cases = np.random.randint(500, 2000, size=10)
cumulative_cases = np.cumsum(new_cases)
vaccinated_population = np.linspace(10, 50, num=10)  # Percentage of vaccinated people
unvaccinated_population = 100 - vaccinated_population
age_groups = ["0-18", "19-30", "31-50", "51-70", "70+"]
vaccination_rates = np.random.randint(30, 90, size=len(age_groups))

# Create figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("COVID-19 Cases & Vaccination Dashboard", fontsize=16)

# Line Plot: New Cases vs. Cumulative Cases
ax1 = axs[0, 0]
ax1.plot(days, new_cases, label="New Cases", marker="o", color="red")
ax1.plot(days, cumulative_cases, label="Cumulative Cases", linestyle="--", color="black")
ax1.set_title("COVID-19 Cases Over Time")
ax1.set_xlabel("Days")
ax1.set_ylabel("Number of Cases")
ax1.legend()
ax1.grid()

# Stacked Area Chart: Vaccinated vs. Unvaccinated Population
ax2 = axs[0, 1]
ax2.fill_between(days, vaccinated_population, color="green", alpha=0.6, label="Vaccinated")
ax2.fill_between(days, 100, vaccinated_population, color="gray", alpha=0.4, label="Unvaccinated")
ax2.set_title("Vaccination Coverage Over Time")
ax2.set_xlabel("Days")
ax2.set_ylabel("Population Percentage")
ax2.legend()

# Bar Chart: Vaccination Rates by Age Group
ax3 = axs[1, 0]
ax3.bar(age_groups, vaccination_rates, color="blue", alpha=0.7)
ax3.set_title("Vaccination Rates by Age Group")
ax3.set_xlabel("Age Group")
ax3.set_ylabel("Vaccination Rate (%)")

# Empty subplot for alignment
axs[1, 1].axis("off")

# Adjust layout and display
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
