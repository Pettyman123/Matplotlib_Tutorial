Task: Stock Market Data Visualization
Create a stock market dashboard to analyze the trends of a company's stock prices over time. Your task includes:

A candlestick-style chart simulating stock price movement (Open, High, Low, Close).
A moving average trend line to smooth fluctuations.
A bar chart for daily trading volume to observe buying/selling patterns.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random stock market data
np.random.seed(42)
days = np.arange(1, 31)  # 30 days of data
open_price = np.random.uniform(100, 200, len(days))
close_price = open_price + np.random.uniform(-10, 10, len(days))
high_price = np.maximum(open_price, close_price) + np.random.uniform(0, 5, len(days))
low_price = np.minimum(open_price, close_price) - np.random.uniform(0, 5, len(days))
volume = np.random.randint(5000, 20000, len(days))

# Calculate 7-day moving average
moving_avg = np.convolve(close_price, np.ones(7)/7, mode='valid')

# Create figure and subplots
fig, axs = plt.subplots(2, 1, figsize=(12, 8), gridspec_kw={'height_ratios': [2, 1]})
fig.suptitle("Stock Market Analysis", fontsize=16)

# Candlestick-style Stock Price Chart
ax1 = axs[0]
ax1.plot(days, open_price, marker='o', linestyle="dotted", color="black", alpha=0.7, label="Open Price")
ax1.plot(days, close_price, marker='o', linestyle="-", color="blue", alpha=0.8, label="Close Price")
ax1.fill_between(days, high_price, low_price, color="gray", alpha=0.3, label="High-Low Range")
ax1.plot(days[3:-3], moving_avg, color="red", linewidth=2, linestyle="dashed", label="7-Day Moving Avg")

ax1.set_title("Stock Price Movement")
ax1.set_xlabel("Days")
ax1.set_ylabel("Price ($)")
ax1.legend()
ax1.grid(True)

# Bar Chart for Trading Volume
ax2 = axs[1]
ax2.bar(days, volume, color="green", alpha=0.6)
ax2.set_title("Daily Trading Volume")
ax2.set_xlabel("Days")
ax2.set_ylabel("Volume")
ax2.grid(axis="y", linestyle="--", alpha=0.5)

# Adjust layout and display
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
