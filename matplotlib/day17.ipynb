Task: Climate Change Analysis Dashboard
Today's task involves visualizing multiple aspects of climate change, focusing on temperature, CO₂ emissions, and sea level rise. You'll work with subplots, dual y-axes, and advanced styling.

Tasks to Complete:
1. Temperature Anomalies Over Time: Use a line chart to display global temperature rise.
2. CO₂ Emissions vs Temperature: Use a scatter plot with regression line to analyze correlation.
3. Sea Level Rise Trends: Use a bar and line combo plot to track the rise in sea level.
4. Polar Ice Cap Reduction: Use a pie chart to represent ice loss over decades.

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(43)
years = np.arange(1980, 2025)
temp_anamolies = np.random.uniform(0.1, 1.5, len(years)) + np.linspace(0, 1.2, len(years))
cos_emissions = np.random.uniform(200,500, len(years)) + np.linspace(0, 100, len(years))
sea_levels = np.random.uniform(0, 5, len(years)) + np.linspace(0, 8, len(years))
```

```python
decades = [ "1980s", "1990s", "2000s", " 2010s", "2020s"]
ice_caps = [100, 85, 70, 50, 20]

fig, axs = plt.subplots(2,2, figsize=(14,14))
fig.suptitle("Climate change", fontweight = "bold")

ax1 = axs[0,0]
ax1.plot(years, temp_anamolies, color="red", linestyle= '-', marker='o', alpha=0.8, label = 'Temp')
ax1.set_title('Temperature Anomalies')
ax1.set_xlabel('Years')
ax1.set_ylabel('Temperature Anomalies')
ax1.grid(True)
ax1.legend()

ax2 = axs[0,1]
sns.regplot(x = cos_emissions, y=temp_anamolies, ax = ax2, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
ax2.set_xlabel("Co2 emission ppm")
ax2.set_ylabel("temp")
ax2.set_title('Temperature vs Co2 Emission')

ax3 = axs[1,0]
ax3.bar(years, sea_levels, color= "skyblue",  alpha = 0.7, label= "sea level rise")
ax3.plot(years, sea_levels, color="darkblue", linewidth=2, label= "Trend line")
ax3.set_title("Global sea level rise in decades")
ax3.set_xlabel("Year")
ax3.set_ylabel("Sea level in m")
ax3.grid(True)
ax3.legend()

ax4 = axs[1,1]
ax4.pie(ice_caps, labels=decades, autopct="%1.1f%%", colors = ["blue", "cyan", "lightblue", "green", "yellow"], startangle=140)
ax4.set_title("Polar ice cap reduction over decades")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
