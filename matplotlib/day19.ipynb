Todayâ€™s challenge involves working with climate and weather data. You will analyze temperature variations, rainfall patterns, and global warming trends using advanced Matplotlib techniques.

Tasks to Complete:
Temperature Trends (Line Plot): Visualize yearly temperature variations.
Rainfall Distribution (Box Plot): Analyze seasonal rainfall patterns.
Global Warming Effect (Heatmap): Show temperature anomalies across years.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Simulate climate data
np.random.seed(42)
years = np.arange(1980, 2023)  # 43 years of data
temperature = np.random.normal(loc=14, scale=2, size=len(years)) + np.linspace(-0.5, 2, len(years))  # Rising trend
rainfall = np.random.normal(loc=1000, scale=200, size=(4, len(years)))  # Seasonal rainfall data (4 seasons)

# Simulate temperature anomalies for heatmap (43 years, 12 months)
temperature_anomalies = np.random.randn(len(years), 12) * 0.5 + np.linspace(-0.5, 2, len(years)).reshape(-1, 1)

# Create figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Climate & Weather Analysis", fontsize=16, fontweight="bold")

# Line Chart: Temperature Trends Over Time
ax1 = axs[0, 0]
ax1.plot(years, temperature, marker="o", linestyle="-", color="red", label="Avg Temperature")
ax1.set_xlabel("Year")
ax1.set_ylabel("Temperature (Â°C)")
ax1.set_title("Global Temperature Trends")
ax1.grid(True)
ax1.legend()

# Box Plot: Rainfall Distribution Per Season
ax2 = axs[0, 1]
rainfall_df = pd.DataFrame(rainfall.T, columns=["Winter", "Spring", "Summer", "Autumn"])
sns.boxplot(data=rainfall_df, ax=ax2, palette="coolwarm")
ax2.set_ylabel("Rainfall (mm)")
ax2.set_title("Seasonal Rainfall Distribution")

# Heatmap: Temperature Anomalies Over Time
ax3 = axs[1, 0]
sns.heatmap(temperature_anomalies, cmap="coolwarm", xticklabels=np.arange(1, 13), yticklabels=years, ax=ax3)
ax3.set_xlabel("Month")
ax3.set_ylabel("Year")
ax3.set_title("Temperature Anomalies Heatmap")

# Adjust layout and display
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
